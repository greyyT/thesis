```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "../data/candidates.csv"

df = pd.read_csv(file_path)
print(f"Shape: {df.shape}")
df.head()
```

```python
# draw a pie chart to visualize the column 'actual_category'

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='actual_category', order=df['actual_category'].value_counts().index)
plt.title('Distribution of Actual Categories')
plt.xlabel('Actual Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# for each distinct value in 'actual_category', filter the DataFrame and save it to a new CSV file
distinct_categories = df['actual_category'].unique()
for category in distinct_categories:
    df_filtered = df[df['actual_category'] == category]
    output_file_path = f"../data/{category}_candidates.csv"
    df_filtered.to_csv(output_file_path, index=False)
    print(f"Filtered data for {category} saved to {output_file_path}")
```
