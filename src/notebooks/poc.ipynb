```python
file_path = "unified_resume_predictions.csv"

df3 = pd.read_csv(file_path)
print(f"Shape: {df3.shape}")
df3.head()
```

```python
# remove column 'source_dataset'
df3 = df3.drop(columns=['source_dataset'], errors='ignore')

df3.head()
```

```python
# create another unified index, then drop 'unified_id' and 'resume_id'
df3['id'] = df3.index + 1
df3 = df3.drop(columns=['unified_id', 'resume_id'], errors='ignore')

df3.head()
```

```python
# save the updated DataFrame to a new CSV file
output_file_path = "candidates.csv"
df3.to_csv(output_file_path, index=False)
```
